```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from citipy import citipy 
import random
import requests
import json
import time
```

```python
# Get random list of 500 cities out of citipy and make dataframe:

cities = []
lats = []
longs = []
unique_lats=set()

for x in range(0,500):
    lat = round(random.uniform(-90, 90),4)
    unique_lats.add(lat)
    while lat in unique_lats:
        lat = round(random.uniform(-90,90),4)
        lats.append(lat)    

unique_longs=set()

for x in range(0,500):
    long = round(random.uniform(-180, 180),2)
    unique_longs.add(long)
    while long in unique_longs:
        long = round(random.uniform(-180, 180),2)
        longs.append(long)

weather_data = pd.DataFrame(list(zip(lats,longs)),columns=['Latitude','Longitude'])
weather_data
    
for index, row in weather_data.iterrows():
    city = citipy.nearest_city(row['Latitude'], row['Longitude']).city_name
    cities.append(city)
    
weather_data['City'] = cities
weather_data.to_csv("coordinates.csv")
```

```python
# Initiate api calls for weather data for each city 

cities_list = enumerate(cities, 1)

for city in cities_list:
    print(f"Requesting city number and name {city}")
    
                            
```

```python
weather_data['Temp'] = ''
weather_data['Humidity'] = ''
weather_data['Wind Speed'] = ''
weather_data['Cloud Cover'] = ''
weather_data['Maximum Temp'] = ''
weather_data['Latitude'] = ''
weather_data['Longitude'] = ''
weather_data['Country'] = ''
weather_data.head()
```

```python
from config import key
key=key

for index,row in weather_data.iterrows():
    time.sleep(1)
    city = row['City']
    url = (f'https://api.openweathermap.org/data/2.5/weather?q={city}&units=imperial&appid={key}')

    response = requests.get(url).json() 
    print(json.dumps(response,indent=4))

    if response['cod'] == 200:
        weather_data.loc[index, 'Temp'] = response['main']['temp']
        weather_data.loc[index, 'Maximum Temp'] = response['main']['temp_max']
        weather_data.loc[index, 'Humidity'] = response['main']['humidity']
        weather_data.loc[index, 'Cloud Cover'] = response['clouds']['all']
        weather_data.loc[index, 'Wind Speed'] = response['wind']['speed']
        weather_data.loc[index, 'Latitude'] = response['coord']['lat']
        weather_data.loc[index, 'Longitude'] = response['coord']['lon']
        weather_data.loc[index, 'Country'] = response['sys']['country']

    else:
        print(f"{city} not found")
        
weather_data.head()
```

```python
# Clean the data, make a dataframe and also export into csv file; make html file of data frame for use in web site.

weather_data_clean = weather_data.dropna().reset_index()
weather_data_clean = weather_data_clean[["City", "Country","Latitude", "Longitude", "Temp", "Maximum Temp" ,"Humidity", "Wind Speed", "Cloud Cover"]]
weather_data_clean.to_csv('weather_data.csv')
weather_data_clean.to_html('weather_data.html')
weather_data_clean.head()
```

```python
# Plotting temperature v. latitude

# Convert data to numeric
weather_data['Temp']=weather_data['Temp'].apply(pd.to_numeric,downcast='float',errors='ignore')
weather_data['Latitude']=weather_data['Latitude'].apply(pd.to_numeric,downcast='float',errors='ignore')
weather_data['Longitude']=weather_data['Longitude'].apply(pd.to_numeric,downcast='float',errors='ignore')
weather_data['Maximum Temp']=weather_data['Maximum Temp'].apply(pd.to_numeric,downcast='float',errors='ignore')
weather_data['Wind Speed']=weather_data['Wind Speed'].apply(pd.to_numeric,downcast='float',errors='ignore')
weather_data['Cloud Cover']=weather_data['Cloud Cover'].apply(pd.to_numeric,downcast='float',errors='ignore')
weather_data['Humidity']=weather_data['Humidity'].apply(pd.to_numeric,downcast='float',errors='ignore')

weather_data.dtypes
```

```python
temp = weather_data['Temp']
lat = weather_data['Latitude']

plt.scatter(lat, temp, marker="o", facecolors="b", edgecolors="b",
            alpha=0.75)
plt.xlim(-90,90)
plt.ylim(-20,120)
plt.title("Temperature by Latitude, 9/14/2020")
plt.xlabel("Latitude")
plt.ylabel("Temperature")

plt.savefig("assets/images/temp_scatter.png")
plt.show()
```

This chart, which arranges the temperatures of cities around the world according to their latitude, clearly shows temperatures rising the closer one is to the equator.

No clear pattern appears here, suggesting little relationship between humidity and latitude (or, by extension, given the previous chart, between hmidity and temperature.  Overall, world humidity readings (at least on land) are predominantly above 60%.

```python
#Linear Regression Calculations and Plots By Hemisphere

import scipy.stats as st
```

```python
x_values = weather_data['Latitude'][weather_data['Latitude'] >= 0]
y_values = weather_data['Maximum Temp'][weather_data['Latitude'] >= 0]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept

print(f'''Pearson's coefficient value is {round(rvalue,2)}. 
The rsquared value = {round(rvalue**2,2)}.''')

x_axis = np.arange(-90, 90, 10)
plt.plot(x_values,regress_values)
plt.scatter(x_values, y_values, marker="+", facecolors="purple", edgecolors="purple",
            alpha=0.75)
plt.ylim(0,120)
plt.title("Maximum Temperatures in Northern Hemisphere, 9/14/2020\n")
plt.xlabel("Latitude")
plt.ylabel("Maximum Temperatures (F)")

plt.savefig("assets/images/max_temp_north.png")
plt.show()

```

```python
x_values = weather_data['Latitude'][weather_data['Latitude'] < 0]
y_values = weather_data['Maximum Temp'][weather_data['Latitude']<0]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept

print(f'''Pearson's coefficient value is {round(rvalue,2)}. 
The rsquared value = {round(rvalue**2,2)}.''')

x_axis = np.arange(-90, 90, 10)
plt.plot(x_values,regress_values)
plt.scatter(x_values, y_values, marker="+", facecolors="orange", edgecolors="orange",
            alpha=0.75)
plt.ylim(0,120)
plt.title("Maximum Temperatures in Southern Hemisphere, 9/14/2020\n")
plt.xlabel("Latitude")
plt.ylabel("Maximum Temperatures (F)")

plt.savefig("assets/images/max_temp_south.png")
plt.show()
```

These charts and those that follow essentially divide the first four charts in half.  So here we have confirmation of the strong correlation--as evidenced by the r and rsquared values--between temperature and distance from the equator. In the southern hemisphere the correlation is positive, meaning that as latitude increases, so does temperature.  The northern hemisphere is reversed, with temperatures lowering as latitude increases.
